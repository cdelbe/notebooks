## Import needed packages and define constants

```python
import requests

URL='https://tada-api.azurewebsites.net/endpoint1/1'
```

## Define a custom exception class

```python
class ApiError(Exception):
    """An API Error Exception"""

    def __init__(self, status):
        self.status = status

    def __str__(self):
        return "ApiError: status={}".format(self.status)
```

## Test GET

```python
response = requests.get(URL)
if response.status_code != 200:
    # This means something went wrong.
    raise ApiError('GET /endpoint1/1 {}'.format(response.status_code))
print(response.json())
```

## Test POST

```python
response = requests.post(URL, json={"name":"Charles"})
if response.status_code != 200:
    raise ApiError('POST /endpoint1/1 {}'.format(response.status_code))
print(response.json())
```
